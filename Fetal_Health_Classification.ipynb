Data preprocessing: oversample + PCA

```python
# Split data:
from sklearn.model_selection import train_test_split
x = data.drop(['fetal_health'], axis = 1)
y = data['fetal_health']
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2, random_state = 123)
```

```python
# Oversampling of minortiy class: https://machinelearningmastery.com/random-oversampling-and-undersampling-for-imbalanced-classification/
# may cause overfitting problem: increases # of minority class = # of majority class
from imblearn.over_sampling import RandomOverSampler

oversample = RandomOverSampler(sampling_strategy= 'not majority')
x_over, y_over = oversample.fit_resample(x_train,y_train)

# before resampling visualization vs. after 
fig,ax = plt.subplots(1,2,figsize = (20,6))
sns.countplot(data = pd.DataFrame(y_train), x = 'fetal_health', ax = ax[0]).set(title='Before Resampling')
sns.countplot(data = pd.DataFrame(y_over), x = 'fetal_health', ax = ax[1]).set(title='After resampling')
```

Optional: 

PCA: Reduce dimentionality and reduce high correlated variable

* In order not to bias result, we do PCA only for training data. And use the reduced matrix U to get the project of testing data.

    * Z_train = U * X_train
    * Z_test = U * X_test


```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_over)
x_test_scaled = scaler.transform(x_test)

from sklearn.decomposition import PCA
pca = PCA(n_components= 0.95)
x_train_pca = pca.fit_transform(x_train_scaled)
x_test_pca = pca.transform(x_test_scaled) # shape:(426, 13)
x_train_pca # shape: (3957, 13)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import RandomizedSearchCV
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, accuracy_score, recall_score, precision_score, f1_score, confusion_matrix, mean_absolute_error, r2_score,mean_squared_error, mean_absolute_percentage_error
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
from xgboost import XGBClassifier
import warnings
warnings.filterwarnings('ignore')
```

```python
def results(y_true, y_pred):
    cm = confusion_matrix(y_true,y_pred)
    acc = accuracy_score(y_true, y_pred)
    #prec = precision_score(y_true, y_pred, average = 'macro') # average = 'macro' calculate metrics for each label
    #rec = recall_score (y_true, y_pred, average = 'macro')
    #f1 = f1_score(y_true, y_pred, average = 'macro')
    print(f'Accuracy score:' , '{:.2%}'.format(acc))
    sns.heatmap(cm, annot = True)
    print(classification_report(y_true, y_pred))

    #print(f'F_1 score:','{:.2%}'.format(f1))

```

* Logistic regression

```python
# Logistic regression: oversample train data + pca
param_grid = {'C':[0.001,0.01,0.1,1,10,100], 'max_iter':[n for n in range(50,1000,100)],'penalty' :['l2','l1']}
log = LogisticRegression(random_state = 42)
clf = RandomizedSearchCV(log, param_grid, cv=5)
clf_log = clf.fit(x_train_pca, y_over)
#clf_log.best_estimator_ 
#clf_log.best_score_
y_pred = clf.predict(x_test_pca)
results(y_test, y_pred)
```

**Evalation of models according to valiadation accuracy**

```python
classifier_list = []
clf_name_list = ['','']

for clf in classifier_list:
    clf.fit(x_over, y_over)
```

```python
train_acc_list = []
test_acc_list = []

for clf, name in zip(classifier_list, clf_name_list):
    y_pred_train = clf.predict(x_over)
    y_pred_test = clf.predict(x_test)
    print(f'Using model:{name}')
    print(f'Acc Train: {accuracy_score(y_over, y_pred_train)}')
    print(f'Acc Test: {accuracy_score(y_test, y_pred_test)}')
    train_acc_list.append(accuracy_score(y_over, y_pred_train))
    test_acc_list.append(accuracy_score(y_test, y_pred_test))
    print('======================================================')
```

```python
clf_name_list = np.array([clf_name_list]).reshape(7,1)
train_acc_list = np.array([train_acc_list]).reshape(7,1)
df = pd.DataFrame(clf_name_list,columns=["clf_name_list"])
df["train_acc_list"] = train_acc_list
sns.barplot(data=df,y="clf_name_list",x="train_acc_list");
```
